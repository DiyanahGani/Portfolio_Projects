# 5. Practice: Sampling

```python
def compare_pop_sample(pop_df, sample_df, col): 
    pop = pop_df[col].value_counts(normalize = True)
    sample = sample_df[col].value_counts(normalize = True)
    pop_sample = pd.concat([pop, sample], 
                           axis = 0, 
                           keys = ["pop", "sample"]).reset_index()
    sns.catplot(x = "level_1", y = col, 
                hue = "level_0", kind = "bar", 
                data = pop_sample)
```

## 5.1 Choose a sample of size 250 using simple random sampling. Is this a good representative of our data considering costumer churn?

```python
np.random.seed(42)
mysample2 = data.sample(n = 250, replace = False)
```

```python
compare_pop_sample(data, mysample2, "Churn")
```

### Answer
It is a good representative sample of population

## 4.2 Is your sample a good representative of the population regarding `Churn`?

**Hint:**
1. Count the percentage of Churned users in the population and save the result as `pop_churn_pcnt`
2. Count the percentage of Churned users in the sample and save the result as `smp_churn_pcnt`

```python
userChurn = pd.read_sql_query("SELECT DISTINCT Churn, COUNT(Churn) AS 'Users' FROM population GROUP BY Churn", conn)
pop_churn_pcnt = round((float(userChurn[userChurn['Churn'] == 1]['Users']) / float(userChurn[userChurn['Churn'] == 0]['Users'])) * 100, 2) # (64 / 647) * 100
print("The percentage of churned users in the population is: {}%".format(pop_churn_pcnt))
```

```python
userChurn = pd.read_sql_query("SELECT DISTINCT Churn, COUNT(Churn) AS 'Users' FROM sample GROUP BY Churn", conn)
smp_churn_pcnt = round((float(userChurn[userChurn['Churn'] == 1]['Users']) / float(userChurn[userChurn['Churn'] == 0]['Users'])) * 100, 2) # (64 / 647) * 100
print("The percentage of churned users in the sample is: {}%".format(smp_churn_pcnt))
```
